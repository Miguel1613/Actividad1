```python
X = TRAIN[["GarageArea","LotArea","PoolArea","1stFlrSF","2ndFlrSF","TotalBsmtSF"]]
```

```python
Y = TRAIN['SalePrice']
```

```python
X = sm.add_constant(X) 

model = sm.OLS(Y, X).fit()

predictions = model.predict(X) 

 

print_model = model.summary()

print(print_model)
```
